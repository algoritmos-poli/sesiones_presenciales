```python
import csv
```

```python
list_students = []
with open('grades.csv', 'r', encoding='utf-8') as file:
    reader = csv.DictReader(file, delimiter=';')
    
    for row in reader:
        # print(row)
        list_students.append(row)
```

```python
def bubble_sort(arr, key):
    n = len(arr)
    # Recorrer todos los elementos
    for i in range(n):
        # Un indicador para optimizar: si no hay intercambios, la lista ya está ordenada
        swapped = False

        # Los últimos i elementos ya están en su lugar
        for j in range(0, n - i - 1):
            # Comparar e intercambiar
            if arr[j][key] > arr[j + 1][key]:
                arr[j][key], arr[j + 1][key] = arr[j + 1][key], arr[j][key]
                swapped = True

        # Si no hubo intercambios en esta pasada, terminar
        if not swapped:
            break
    return arr
```
