```python
def linSearch(L, e):
    '''
    Determina si un elemento se encuentra o no dentro de un lista

    Parameters:
    L (list): Lista de numeros
    e (int): Elemento a buscar

    Returns:
    bool: Retorna falso o verdadero dependiendo si el numero se encuentra o no en la lista
    '''

    for i in range(len(L)):
        print("*", end = "")
        if L[i] == e:
            print()
            return True
    print()
    return False
```

```python
# Test
L = [5,2,10,1,9]
e1 = 6
v1 = linSearch(L, e1)
print(f"{e1} in {L}: {v1}")
e2 = 2
v2 = linSearch(L, e2)
print(f"{e2} in {L}: {v2}")
```

```python
def ordSearch(L,e):
    '''
    Determina si un elemento se encuentra o no dentro de un lista

    Parameters:
    L (list): Lista de numeros. Esta lista debe estar ordenada
    e (int): Elemento a buscar

    Returns:
    bool: Retorna falso o verdadero dependiendo si el numero se encuentra o no en la lista
    '''
    for i in range(len(L)):
        print("*", end ='')
        if L[i] == e:
            print()
            return True
        elif L[i] > e:
            print()
            return False
    print()
    return False
```

## Algoritmos de busqueda

```python
# Test
L = [1,2,5,9,10]
e1 = 6
v1 = ordSearch(L, e1)
print(f"{e1} in {L}: {v1}")
e2 = 2
v2 = ordSearch(L, e2)
print(f"{e2} in {L}: {v2}")
```

```python
def binSearch(L,e):
    '''
    Determina si un elemento se encuentra o no dentro de un lista

    Parameters:
    L (list): Lista de numeros. Esta lista debe estar ordenada
    e (int): Elemento a buscar

    Returns:
    bool: Retorna falso o verdadero dependiendo si el numero se encuentra o no en la lista
    '''
    low = 0
    high = len(L) - 1
    while low <= high:
        print("*", end="")
        mid = (high + low)//2
        if e > L[mid]:
            low = mid + 1
        elif e < L[mid]:
            high = mid - 1
        else:
            print()
            return True
    print()
    return False
```

```python
# Test
L = [1,2,5,9,10]
e1 = 6
v1 = binSearch(L, e1)
print(f"{e1} in {L}: {v1}")
e2 = 2
v2 = binSearch(L, e2)
print(f"{e2} in {L}: {v2}")
```

```python
def selectionSort(L):
    start_L = 0
    end_L = len(L)
    print(L)
    while start_L < end_L:
        sel = L[start_L]
        for i in range(start_L + 1,end_L):
            if(sel > L[i]):
                tmp = sel
                sel = L[i]
                L[i] = tmp
        L[start_L] = sel
        print(L)
        start_L += 1
```

```python
# Test
a = [3, 0, 1, 8, 7, 2, 5, 4, 9, 6]
selectionSort(a)
```

```python
def leer_def(nombre_a):

    traductor = {}

    db_def = open(nombre_a,encoding = 'utf-8')
    traductor = {}  # Suponiendo que esto es un diccionario donde estás almacenando las traducciones
    for linea in db_def:
        # Verificar si la línea contiene el carácter ';'
        if ';' in linea:
            espanol, comando = linea.strip().split(';')
            traductor[espanol] = comando
        else:
            print(f"Error: La línea no contiene ';' - {linea}")

    return traductor

traductor = leer_def("datos.txt")
```

```python
def escribir_def(traductor,nombre_a):
    texto = ''
    for clave in traductor:
        texto += clave +'; '+ ', '.join(traductor[clave]) + '\n'

    db_def = open(nombre_a,'w',encoding='utf-8')
    db_def.write(texto)
    db_def.close()
```
